## How do perceptions of UPenn and UIUC differ across platforms

Loading Data

```python
sentiment_df = pd.read_csv('/Commjhub/jupyterhub/home/emch06/comm3180_S26/comm3180_G3_Data_Project/data/raw/Overall_data/upenn_uiuc_sentiments.csv')
```

### Overall Sentiment by School 

```python
sentiment_df.groupby("school")["manual_sentiment_score"].mean()
```

```python
sentiment = sentiment_df.groupby("school")["manual_sentiment_score"].mean()

sentiment.plot(kind="bar")

plt.axhline(0) 
plt.title("Sentiment Comparison: UPenn vs UIUC")
plt.ylabel("Sentiment Score")
plt.xticks(rotation=0)

plt.show()
```

# Overall

UPenn outperforms on financial outcomes (better debt-to-income ratio), yet its community expresses more negative sentiment. Meanwhile, UIUC has a heavier debt burden but a more positive community outlook.

-Interesting to consider is how UPenn has a more negative sentiment during school, but a more positive sentiment post-grad. 

-Upenn students still report significantly higher stress, lower belonging, and lower school spirit during their time enrolled, yet their ROI is substantially higher post-graduation. UIUC students report higher happiness and school spirit while in school, but lower long-term financial return. Is there a tradeoff between wellbeing during school and financial outcomes after graduation, or is it just because UPenn is an Ivy and UIUC a public in-state.
