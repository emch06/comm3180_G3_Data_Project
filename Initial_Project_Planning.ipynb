## Initial Project Plan

For our data analysis on both UIUC and Penn, we have divided it into 2 categories: 
- Quantitative (median salary, debt, ROI, demographics, etc)  
- Qualitative (sentiment analysis) 

Research Question: **How much does institutional prestige actually translate into measurable post-graduate advantage?**
- Another way to percieve this question: Which school has an overall, stronger return on investment according to post-grad salary, career progression, and school cost? (Engineering, Business, and Overall)

**Quantitative** 
- median income levels
- overall mobility index 
- average salaries for people in that industry (engineering & business and overall) 
- ROI (compare median income levels and tuition → seeing if actually worth it) 
- median debt
- geography of alumni
  
**Qualitative** (What are the overall sentiments for each school regarding quality-of-life, happiness levels, and general media impressions of hire-ability?)
- sentiment analysis 
- school association 
- resources: reddit/social media/news articles discussions/Linkedin, happiness levels of post-grads, overall sentiments, survey data from Daily Penn or UIUC newspaper, alumni rating of undergraduate experience --> combine to form a sentiment analysis
